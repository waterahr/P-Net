```python
def load_img(path, image_shape=(299, 299)):
    img = image.load_img(path, target_size=(image_shape[0], image_shape[1], 3))
    plt.imshow(img)
    plt.show()
    img = image.img_to_array(img)
    img = img[np.newaxis, ...]
    return img

load_img(X_train[0]).shape
```

# PartPool Model Summary

```python
whole = [1,2,3,4,5,16,34]
hs = [0,8,20,21,25,28,36,37,44,54]
ub = [15,19,23,27,30,39,40,46,50,51,55,56,58,59,60]
lb = [10,14,18,22,24,29,31,32,45,47,53,57]
sh = [9,26,42,43,48,49]
at = [6,7,11,12,13,17,33,35,38,41,52]

titles = ["whole", "hs", "ub", "lb", "sh", "at"]
```

```python
"""
label_acc_dual = pickle.load(open("./PETA/PartPoolInception/thr3_v8_epoch120_valloss0.984745.hdf5_results.file", "rb"))['label_acc']
label_acc_spatial = pickle.load(open("./PETA/PartPoolInception/v1_epoch048_valloss0.979095.hdf5_results.file", "rb"))['label_acc']
label_acc_channel = pickle.load(open("./PETA/PartPoolInception/v7_v7_epoch100_valloss0.952320.hdf5_results.file", "rb"))['label_acc']
"""
version = "v8"
model = build_partpool_inception([len(whole), len(hs), len(ub), len(lb), len(sh), len(at)], version)
```

# Define Attention Output Model

```python
model.load_weights("../../models/PETA/PartPoolInception/v8_epoch120_valloss0.984745.hdf5", by_name=True)
```

```python
layer_name = ["lambda_17", "lambda_20", "lambda_23", "lambda_26", "lambda_29", "lambda_32"]
model_refined_feature = Model(inputs=model.input, outputs=[model.get_layer(layer_name[i]).output for i in range(6)])
model_refined_feature.summary()
```

```python
#layer_name = ["lambda_16", "lambda_19", "lambda_22", "lambda_25", "lambda_28", "lambda_31"]
layer_name = ["dense_2", "dense_5", "dense_8", "dense_11", "dense_14", "dense_17"]
model_attention = Model(inputs=model.input, outputs=[model.get_layer(layer_name[i]).output for i in range(6)])
model_attention.summary()
```
